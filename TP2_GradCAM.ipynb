# Grad-CAM 
* **Overview:** Given an image, and a category (‘tiger cat’) as input, we forward-propagate the image through the model to obtain the `raw class scores` before softmax. The gradients are set to zero for all classes except the desired class (tiger cat), which is set to 1. This signal is then backpropagated to the `rectified convolutional feature map` of interest, where we can compute the coarse Grad-CAM localization (blue heatmap).

## Part 1: Grad-CAM implementation

A hook in backward or forward is a function to add to a tensor that will be executed in backward or forward 

```python
# Your code here
def hook_fn(module: torch.nn.Module, input: torch.Tensor, output: torch.Tensor):
    """
    hook function to get the grad
    """
    
    for grad in output:
        print("-- Output Grad ---")
        print(grad.shape)
    print("\n")
    

def show_grad_cam(image: torch.Tensor) -> None :
    """
    show grad cam

    Parameters:
    -----------
    image:  torch.Tensor

    Returns:
    --------
    None
    """
    # hook
    features_layer = densenet121.features
    features_layer.register_full_backward_hook(hook_fn)
    classifier_layer = densenet121.classifier
    # features_layer.register_full_backward_hook(hook_fn)


    # forward
    feature_map = features_layer(image)
    feature_map = feature_map.mean(axis=[2,3]) # avg
    print(feature_map.shape)
    output = classifier_layer(feature_map)

    values, indices = torch.topk(output, 3)
    top_3_indexes =  indices[0].numpy()
    top_3_classes = [classes[x] for x in indices[0].numpy()]
    # values traversal for the top_3 scores
    for i, value in enumerate(values): 
        print(top_3_classes[i])
        output[0, top_3_indexes[i]].backward()
        # get the gradient of the feature map
        grad = features_layer.weight.grad
        # compute the weights
        weights = grad.mean(axis=[2, 3])
        # compute the grad cam
        grad_cam = F.relu((weights * feature_map).sum(axis=0))
        # normalize the grad cam
        grad_cam = grad_cam - grad_cam.min()
        grad_cam = grad_cam / grad_cam.max()
        # show
        plt.imshow(grad_cam.detach().numpy(), cmap='hot')

    
```
